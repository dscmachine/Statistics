## Skewness & Kurtosis

```python
from scipy.stats import skew 
import numpy as np  
  
x = np.random.normal(0, 2, 10000)   # random values based on a normal distribution 
  
print (x) 
  
print('\nSkewness for data : ', skew(x))
```

```python
import numpy as np
from scipy.stats import kurtosis, skew

x = np.random.normal(0, 2, 10000)   # create random values based on a normal distribution

print(x)
print( '\nkurtosis for data : ',kurtosis(x) )
```

```python
from scipy.stats import skew 
import numpy as np  
import matplotlib.pyplot as plt   
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
plt.plot(x1, y1, '*') 
  
print( '\nSkewness for data : ', skew(y1))
```

```python
from scipy.stats import kurtosis 

import numpy as np  

import pylab as p  

x1 = np.linspace( -5, 5, 1000 ) 

y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 


p.plot(x1, y1, '*') 

print( ' Kurtosis value :', kurtosis(y1))
```
